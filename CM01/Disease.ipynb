```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.set_printoptions(suppress=True)
```

# Disease: Priors, Sensitivity, Specificity

```python
#Generate data showing P(D|T) when T(T), prior variable
def variable_prior():
    
    #Prior varies from 0-1
    prior = np.arange(0, 1.01, .01)
    
    #Apply bayes
    calc = lambda x: (0.95 * x) / (0.95 * x + 0.90 * (1-x))
    post = calc(prior)
    
    #Plot
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(prior, post, color='#AA0000', linestyle='dashed', marker='o')
    ax.set_xlabel('Prior', fontsize=16)
    ax.set_ylabel('Posterior', fontsize=16)
    ax.set_title('P(D|T) with variable prior', fontsize=20)
    
```

```python
#Generate data showing P(D|T) when T(T), variable sensitivity
def variable_sensitivity():
    
    #Sensitivity varies from 0-1
    sen = np.arange(0, 1.01, .01)
    
    #Apply bayes
    calc = lambda x: (x * 0.01) / (x * 0.01 + 0.90 * 0.99)
    post = calc(sen)
    
    #Plot
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(sen, post, color='#AA0000', linestyle='dashed', marker='o')
    ax.set_xlabel('Sensitivity', fontsize=16)
    ax.set_ylabel('Posterior', fontsize=16)
    ax.set_title('P(D|T) with variable sensitivity', fontsize=20)
    ax.set_ylim(0,1)
    
```

```python
#Generate data showing P(D|T) when T(T), variable specificity
def variable_specificity():
    
    #Specificity varies from 0-1
    spec = np.arange(0, 1.01, .01)
    
    #Apply bayes
    calc = lambda x: (0.95 * 0.01) / (0.95 * 0.01 + x * 0.99)
    post = calc(spec)
    
    #Plot
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.plot(spec, post, color='#AA0000', linestyle='dashed', marker='o')
    ax.set_xlabel('Specificity', fontsize=16)
    ax.set_ylabel('Posterior', fontsize=16)
    ax.set_title('P(D|T) with variable specificity', fontsize=20)
```

```python
variable_prior()
variable_sensitivity()
variable_specificity()
```
